```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#untuk import dataset
df = pd.read_csv('/content/drive/MyDrive/heart_disease.csv')
print(df.head())
df.head(10)#untuk menampilkan preview dataset
```

```python
#untuk mengetahui jumlah baris&kolom
df.shape
```

```python
#untuk mengetahui nama kolom
df.columns
```

```python
#untuk mengetahui tipe data
df.info()
```

```python
#mengecek jumlah missing value awal
print("Jumlah Missing Value Awal:")
print(df.isna().sum())
```

```python
columns_to_drop = ["ca", "slope", "thal","id","dataset"]

# Drop the columns from the dataset
df.drop(columns=columns_to_drop, inplace=True)

# Verify the remaining columns
print("kolom setelah dihapus:")
print(df.columns)
```

```python
#memisahkan nama kolom berdasarkan tipe data
kolom_angka = df.select_dtypes(include=['float64', 'int64']).columns
kolom_teks = df.select_dtypes(include=['object']).columns
```

```python
#mengisi missing value pada kolom angka dengan median
for col in kolom_angka:
    if df[col].isnull().sum() > 0:
        val_median = df[col].median()
        df[col] = df[col].fillna(val_median)
```

```python
#mengisi missing value pada kolom teks dengan modus
for col in kolom_teks:
    if df[col].isnull().sum() > 0:
        val_modus = df[col].mode()[0]
        df[col] = df[col].fillna(val_modus)
```

```python
#mengecek kembali untuk memastikan data bersih
print("Jumlah Missing Value Setelah Diisi:")
print(df.isna().sum())
```

```python
#Statistik deskriptif data angka
print("--- Statistik Numerik ---")
print(df[kolom_angka].describe())
print("\n" + "="*50 + "\n")
```

```python
#Statistik deskriptif data kategori
print("--- Statistik Kategori ---")
print(df[kolom_teks].describe())
```

```python
#mengecek nama kolom asli terlebih dahulu untuk memastikan
print("Nama Kolom Sebelum Diubah:")
print(df.columns.tolist())
```

```python
#mengubah nama fitur menjadi Bahasa Indonesia
nama_baru = {
    'age': 'umur',
    'sex': 'jenis_kelamin',
    'cp': 'tipe_nyeri_dada',
    'trestbps': 'tekanan_darah',
    'chol': 'kolesterol',
    'fbs': 'gula_darah_puasa',
    'restecg': 'hasil_ekg',
    'thalch': 'detak_jantung_maks',
    'thalach': 'detak_jantung_maks',
    'exang': 'angina_olahraga',
    'oldpeak': 'depresi_st',
    'slope': 'kemiringan_st',
    'ca': 'jumlah_pembuluh_utama',
    'thal': 'thalasemia',
    'num': 'diagnosis_penyakit'
}
```

```python
#mengeksekusi perubahan nama
df.rename(columns=nama_baru, inplace=True)
print("Nama kolom berhasil diubah ke Bahasa Indonesia.")
```

```python
#mengubah data text menjadi bahasa Indonesia
mapping_nilai = {
    'jenis_kelamin': {'Male': 'Laki-laki', 'Female': 'Perempuan'},
    'tipe_nyeri_dada': {
        'typical angina': 'angina tipikal',
        'atypical angina': 'angina atipikal',
        'non-anginal pain': 'nyeri non-angina',
        'asymptomatic': 'tanpa gejala'
    },
    'angina_olahraga': {'True': 'Ya', 'False': 'Tidak', True: 'Ya', False: 'Tidak'},
    'hasil_ekg': {
        'normal': 'normal',
        'st-t abnormality': 'abnormalitas st-t',
        'lv hypertrophy': 'hipertrofi lv'
    },
    'kemiringan_st': {'upsloping': 'menaik', 'flat': 'datar', 'downsloping': 'menurun'},
    'thalasemia': {
        'normal': 'normal',
        'fixed defect': 'cacat tetap',
        'reversable defect': 'cacat dapat dipulihkan'
    }
}
```

```python
#menampilkan hasil
for kolom, peta in mapping_nilai.items():
    if kolom in df.columns:
        df[kolom] = df[kolom].replace(peta)
print("\nContoh Data Final:")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#visualisasi
# 1. Distribusi Variabel Target (diagnosis_penyakit)
plt.figure(figsize=(8, 5))
sns.countplot(x='diagnosis_penyakit', data=df, palette='viridis')
plt.title('Distribusi Diagnosis Penyakit')
plt.xlabel('Diagnosis Penyakit (0: Tidak Ada, >0: Ada Penyakit)')
plt.ylabel('Jumlah Pasien')
plt.show()
```

```python
# 2. Hubungan antara jenis_kelamin dan diagnosis_penyakit
plt.figure(figsize=(10, 6))
sns.countplot(x='jenis_kelamin', hue='diagnosis_penyakit', data=df, palette='magma')
plt.title('Diagnosis Penyakit berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Pasien')
plt.legend(title='Diagnosis Penyakit')
plt.show()
```

```python
# 3. Distribusi Umur
plt.figure(figsize=(10, 6))
sns.histplot(df['umur'], bins=15, kde=True, color='skyblue')
plt.title('Distribusi Umur Pasien')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# 4. Hubungan antara tipe_nyeri_dada dan diagnosis_penyakit
plt.figure(figsize=(12, 6))
sns.countplot(x='tipe_nyeri_dada', hue='diagnosis_penyakit', data=df, palette='viridis')
plt.title('Diagnosis Penyakit berdasarkan Tipe Nyeri Dada')
plt.xlabel('Tipe Nyeri Dada')
plt.ylabel('Jumlah Pasien')
plt.legend(title='Diagnosis Penyakit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 5. Distribusi Tekanan Darah (trestbps) berdasarkan Diagnosis Penyakit
plt.figure(figsize=(12, 6))
sns.boxplot(x='diagnosis_penyakit', y='tekanan_darah', data=df, palette='plasma')
plt.title('Distribusi Tekanan Darah berdasarkan Diagnosis Penyakit')
plt.xlabel('Diagnosis Penyakit (0: Tidak Ada, >0: Ada Penyakit)')
plt.ylabel('Tekanan Darah')
plt.show()
```

```python
# 6. Distribusi Kolesterol (chol) berdasarkan Diagnosis Penyakit
plt.figure(figsize=(12, 6))
sns.boxplot(x='diagnosis_penyakit', y='kolesterol', data=df, palette='mako')
plt.title('Distribusi Kolesterol berdasarkan Diagnosis Penyakit')
plt.xlabel('Diagnosis Penyakit (0: Tidak Ada, >0: Ada Penyakit)')
plt.ylabel('Kolesterol')
plt.show()
```
